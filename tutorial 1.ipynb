```python
primes = []

for num in range(1, 1001):
    if num > 1:  # Prime numbers are greater than 1
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            primes.append(num)

print(primes)
```

```python
n = int(input("Enter a natural number: "))
count = 0

if n == 0:
    count = 1
else:
    while n != 0:
        count += 1
        n //= 10

print("Number of digits:", count)
```

```python
class BankAccount:
    def __init__(self, account_number, initial_balance):
        self.account_number = account_number
        self.balance = initial_balance
        self.transaction_history = []

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            self.transaction_history.append(f"Deposited ${amount}")
            print(f"Deposited ${amount}. Current balance: ${self.balance}")
        else:
            print("Invalid deposit amount.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            self.transaction_history.append(f"Withdrew ${amount}")
            print(f"Withdrew ${amount}. Current balance: ${self.balance}")
        else:
            print("Invalid withdrawal amount or insufficient balance.")

    def display_balance(self):
        print(f"Account Number: {self.account_number}")
        print(f"Current Balance: ${self.balance}")

    def apply_interest(self, rate):
        interest = self.balance * (rate / 100)
        self.balance += interest
        self.transaction_history.append(f"Interest applied. Added ${interest} as interest.")
        print(f"Interest applied. Current balance: ${self.balance}")

    def display_transaction_history(self):
        print("Transaction History:")
        for transaction in self.transaction_history:
            print(transaction)


# Example usage:
account = BankAccount("1234567890", 1000)
account.deposit(500)
account.withdraw(200)
account.display_balance()
account.apply_interest(2.5)
account.display_transaction_history()
```

```python
import datetime

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def display_available_books(self):
        print("Available Books:")
        for book in self.books:
            if book.is_available():
                print(f"{book.title} by {book.author}")

    def borrow_book(self, book, member):
        if book.is_available():
            due_date = datetime.date.today() + datetime.timedelta(days=14)
            book.borrow(member, due_date)
            print(f"{member.full_name} has borrowed {book.title}. Due date: {due_date}")
        else:
            print(f"{book.title} is not available for borrowing.")

    def return_book(self, book):
        book.return_book()
        print(f"{book.title} has been returned.")


class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True
        self.borrower = None
        self.due_date = None

    def is_available(self):
        return self.available

    def borrow(self, member, due_date):
        self.available = False
        self.borrower = member
        self.due_date = due_date

    def return_book(self):
        self.available = True
        self.borrower = None
        self.due_date = None


class Member:
    def __init__(self, full_name, member_id):
        self.full_name = full_name
        self.member_id = member_id


# Example usage:
library = Library()

book1 = Book("The Great Gatsby", "F. Scott Fitzgerald")
book2 = Book("To Kill a Mockingbird", "Harper Lee")
book3 = Book("Pride and Prejudice", "Jane Austen")

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

member1 = Member("John Smith", "M001")
member2 = Member("Jane Doe", "M002")

library.display_available_books()

library.borrow_book(book2, member1)
library.borrow_book(book3, member2)

library.display_available_books()

library.return_book(book2)

library.display_available_books()
```

```python
import numpy as np

def generate_random_matrix(N):
    matrix = np.random.randint(1, 10, (N, N))
    return matrix

def calculate_row_sums(matrix):
    row_sums = np.sum(matrix, axis=1)
    return row_sums

def calculate_column_sums(matrix):
    column_sums = np.sum(matrix, axis=0)
    return column_sums

# Get user input for the size of the matrix
N = int(input("Enter the size of the matrix: "))

# Generate random matrix
matrix = generate_random_matrix(N)
print("Random Matrix:")
print(matrix)

# Calculate row sums
row_sums = calculate_row_sums(matrix)
print("Row Sums:")
print(row_sums)

# Calculate column sums
column_sums = calculate_column_sums(matrix)
print("Column Sums:")
print(column_sums)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a sample dataframe with performance metrics
data = {
    'Model': ['Model A', 'Model B', 'Model C'],
    'Accuracy': [0.85, 0.92, 0.88],
    'Precision': [0.82, 0.88, 0.89],
    'Recall': [0.78, 0.85, 0.87]
}
df = pd.DataFrame(data)

# Calculate the mean and standard deviation of each metric across all models
mean_values = df.mean()
std_values = df.std()

# Visualize the mean performance of each metric using a bar chart
sns.barplot(x=mean_values.index, y=mean_values.values)
plt.xlabel('Metrics')
plt.ylabel('Mean Values')
plt.title('Mean Performance of Metrics')
plt.show()

# Plot error bars on the bar chart to represent the standard deviation
sns.barplot(x=mean_values.index, y=mean_values.values, yerr=std_values.values, capsize=5)
plt.xlabel('Metrics')
plt.ylabel('Mean Values')
plt.title('Mean Performance of Metrics with Error Bars')
plt.show()

# Compare the performance of each model on a specific metric using a line plot
metric = 'Precision'
sns.lineplot(x='Model', y=metric, data=df)
plt.xlabel('Models')
plt.ylabel(metric)
plt.title(f'Performance of Models on {metric}')
plt.show()
```
